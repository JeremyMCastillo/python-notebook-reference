```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
boston_dataset = datasets.load_boston()
X_full = boston_dataset.data
Y = boston_dataset.target
print(X_full.shape)
print(Y.shape)
```

# Using SelectKBest to Automatically Select the Most Discriminative Feature
Using the following methods we can get a single feature, thus turning our data set into a vectur that we can test for linearity.

```python
selector = SelectKBest(f_regression, k=1)
selector.fit(X_full, Y)
X = X_full[:, selector.get_support()]
print(X.shape)
```

```python
plt.scatter(X, Y, color='black')
plt.show()
```

# Checking for Linearity
Since the data curves, instead of trending straight, we can be mostly sure this set is nonlinear. Let's pretend the data is linear and see if it fits.

```python
regressor = LinearRegression(normalize=True)
regressor.fit(X, Y)
plt.scatter(X, Y, color='black')
plt.plot(X, regressor.predict(X), color='blue', linewidth=3)
plt.show()
```

# Solve Nonlinear Data with Support Vector Machine
The above data is obviously nonlinear. We can use Support Vector Machine, a nonlinear model to help predict this data.

```python
regressor = SVR()
regressor.fit(X, Y)
plt.scatter(X, Y, color='black')
plt.plot(X, regressor.predict(X), color='blue', linewidth=3)
plt.show()
```

# Random Forest Works Too

```python
regressor = RandomForestRegressor()
regressor.fit(X, Y)
plt.scatter(X, Y, color='black')
plt.scatter(X, regressor.predict(X), color='blue', linewidth=3)
plt.show()
```
